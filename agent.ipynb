```python
#disable tracing before importing agents
os.environ["OPENAI_AGENTS_DISABLE_TRACING"] = "1"
from agents import Agent, Runner, trace , OpenAIChatCompletionsModel

```

```python
#configuring openrouter

Client = AsyncOpenAI(api_key=os.getenv("API_KEY") , 
                     base_url="https://openrouter.ai/api/v1")
```

```python
Model = OpenAIChatCompletionsModel(model="gpt-4o-mini" , 
                                   openai_client=Client,) # this is passed to the model in the agent call

system_prompt = "You are a football ethusiast, you are to talk about football related matters both latest and old"

Agenta = Agent(name="Agartha", 
               instructions=system_prompt, 
               model=Model)
```

```python
user_prompt = input("enter input....")

runner = await Runner.run(Agenta,user_prompt)
print(runner.final_output)
```

```python
from IPython.display import Markdown,display
from pypdf import PdfReader
import gradio as gr
```

```python
#this is to personalised the agent, i didn't really use it but just needed to keep it incase i might need it
pdf_reader = PdfReader("Resources/biography.pdf") # to read pdfs
prof_summary = "" 
for page in pdf_reader.pages:
    text = page.extract_text()
    if text:
        prof_summary += text + "\n"
        print(prof_summary)
```

```python
name = "helen"
system_command = (f"""
you name is{name},a helpful assistant,you help answer users question concerning a hackathon, use the{prof_summary} document provided for reference""")
```

```python
agent_helen = Agent(
    model=Model,
    instructions=system_command,
    name=name
)
```

```python
async def livechat(message,history):
    messages = [{"role": msg["role"], "content": msg["content"]} for msg in history]
    messages.append({"role":"user", "content":message})
    result = await Runner.run(agent_helen,messages)
    return result.final_output
```

```python
gr.ChatInterface(
    livechat,
    title="Helen",
    description="Ask Anything about the hackathon"
).launch()
```
