Agent as a tool

```python
import os
from dotenv import load_dotenv
os.environ["OPENAI_AGENTS_DISABLE_TRACING"] = "1"
from agents import Agent,Runner,function_tool,OpenAIChatCompletionsModel
from openai import AsyncOpenAI
```

```python
load_dotenv()

openai_api_key = os.getenv("OPENAI_API_KEY")

if openai_api_key:
    print("API KEY DETECTED")
else:
    print("API KEY NOT DETECTED")
```

```python
client = AsyncOpenAI(api_key=openai_api_key,base_url="https://openrouter.ai/api/v1")

Model =  OpenAIChatCompletionsModel(openai_client=client,model="gpt-4o-mini")
```

```python
# creating specialist agents

# Agents Instructions
research_system_prompt = """
You research topics thoroughly and return detailed findings.
"""

write_system_prompt = """
You write clear, engaging content based on provided information.
"""
master_system_prompt = """
You manage content creation.
    
    When asked to create content:
    1. Use research_topic to gather information
    2. Use write_content to create the final piece
    3. Review and present the result
"""

# Agents 
researcher = Agent(
    name="Researcher",
    instructions=research_system_prompt,
    model=Model
)
writer = Agent(
    name= "Writer",
    instructions=write_system_prompt,
    model=Model
)

# agent as too function
@function_tool
async def research_topic(topic: str) -> str:
    """
    Research a topic thoroughly.
    
    Args:
        topic: The topic to research
    """
    runner = await Runner.run(researcher,f"research this topic: {topic}")
    return runner.final_output

@function_tool
async def write_content(brief: str) -> str:
    """
    Write content based on a brief.
    
    Args:
        brief: The writing brief with topic and key points
    """
    runner = await Runner.run(writer,brief)
    return runner.final_output


orchestrator = Agent(
    name="Content Manager",
    instructions=master_system_prompt,
    model=Model,
    tools=[]
)



```

```python
user_input = input("Enter Prompt")
result = await Runner.run(
    orchestrator,
    user_input
)
print(result.final_output)
```
